```python
import pandas as pd 
import matplotlib.pyplot as plt 
nGen = 25
for iteration in range(30) :
    dfGraf = pd.read_csv("percobaan_"+str(iteration+1)+"_"+str(nGen)+".csv")
    x = dfGraf["x"].tolist()
    y = dfGraf["y"].tolist()
    plt.title("Hasil 30 kali percobaan")
    plt.xlabel('Diversity', fontsize=15)
    plt.ylabel('Activtity', fontsize=15)
    plt.scatter(y, x)
```
